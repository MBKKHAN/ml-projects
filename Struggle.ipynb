```python
alpha = 0.977400
beta = -0.067270
def calculate_prior(time,alpha,beta):
    p= 1.0 / (1.0 + np.exp(np.dot(beta, time) + alpha))
    return p
time_est=np.linspace(time.min()-5,time.max()+5,1000)[:,None]
sleep_est = calculate_prior(time_est, alpha, beta)
```

```python
plt.plot(time_est,sleep_est,color='green',lw=1,label='Model')
plt.scatter(time,sleep_obs,s=10,edgecolor='navy',alpha=0.2,label='Observation')
plt.legend()
plt.title('Prior Probability Distribution for Sleep as Function of Time')
plt.xticks([-60,-30,0,30,60,90,120],sleep_labels)
plt.xlabel('PM Time')
plt.ylabel('Probability')
```

```python
print('The probability of sleep increases to above 50% at 10:{}%\n'.format(
    int(time_est[np.where(sleep_est>0.5)[0][0]][0])
)
     )
print('The probability of feel asleep at 9:30PM is {:.2f}%'.format(100*calculate_prior(-30,alpha,beta)))
print('The probability of feel asleep at 10:00PM is {:.2f}%'.format(100*calculate_prior(0,alpha,beta)))
print('The probability of feel asleep at 10:30PM is {:.2f}%'.format(100*calculate_prior(30,alpha,beta)))
```

```python
# likelihoods

# P(light | sleep)
light_sleep = 0.01

# P(-light | sleep)
nolight_sleep = 0.99

# P(light | -sleep)
light_nosleep = 0.8

# P(-light | -sleep)
nolight_nosleep = 0.2
```

```python
def update_probability(time_offset,light):
    prior_probability= calculate_prior(time_offset,alpha,beta)
 #0=nolight ,1=light
    if light==0:
        likelihood=nolight_sleep 
        non_likelihood=nolight_nosleep
    elif light==1:
        likelihood=light_sleep
        non_likelihood=light_nosleep

    numerator= (likelihood*prior_probability)
    denominator=(likelihood*prior_probability)+(non_likelihood*(1-prior_probability))
    conditional_probability= numerator/denominator
    if type(time_offset)==int:
        time=datetime.datetime(2017, 1, 1, 10, 0, 0)
        new_time=str((time+datetime.timedelta(minutes=time_offset)).time())

        print('Time:{}PM\nLight is {}'.format((new_time),'ON' if light==1 else 'OFF'))
        print('The Prior Probability of sleep is {:.2f}%'.format(100*prior_probability))
        print('The Updated Probability of sleep is {:.2f}%'.format(100* conditional_probability))

    return  conditional_probability
    
```

```python
result=update_probability(time_offset=14,light=1)
```

```python
result=update_probability(time_offset=14,light=0)
```

```python
light_on_probability=update_probability(time_est,light=1)
light_off_probability=update_probability(time_est,light=0)
```

```python
plt.plot(time_est,sleep_est,color='black',linestyle='--',lw=3,label='Prior Sleep')
plt.plot(time_est,light_on_probability,color='darkred',lw=1,label='Light on Sleep')
plt.plot(time_est,light_off_probability,color='darkgreen',lw=1,label='Light off Sleep')
plt.scatter(time,sleep_obs,edgecolor='blue',alpha=0.1,label='Observation')
plt.legend(loc=0)
plt.xticks([-60,-30,0,30,60,90,120],sleep_labels)
# plt.yticks([0,1],['Awake','Sleep'])
plt.xlabel('Time(PM)')
plt.ylabel('Probability')
plt.show()
```
