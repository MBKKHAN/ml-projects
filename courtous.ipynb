```python
import numpy as np
np.random.seed(1337)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
plt.style.use('ggplot')
```

```python
num_bins=50
num_samples=  10000000
data=100+10*np.random.randn(num_samples,2)
```

```python
joint_prob,edges=np.histogramdd(data,num_bins)
edges=np.array(edges)[:,1:]
```

```python
joint_prob
```

```python
edges
```

```python
joint_prob.shape
```

```python
joint_prob
```

```python
cut=int(num_bins/2)
X_Vals=np.array(joint_prob[cut])
X_Vals=X_Vals/X_Vals.sum()
X_Vals
```

```python
Y_Vals= np.array([row[cut] for row in joint_prob])
Y_Vals = Y_Vals/Y_Vals.sum()
Y_Vals
```

```python
plt.bar(edges[0],X_Vals)
plt.title('Distribution for X Values')
plt.xlabel('X_Values')
plt.ylabel('Frequency')
```

```python
plt.bar(edges[1],Y_Vals)
plt.title('Distribution of Y Values')
plt.xlabel('Y_Values')
plt.ylabel('Frequency')
```

```python
X=edges[0]
Y=edges[1]
X,Y=np.meshgrid(X,Y)
Z=joint_prob/joint_prob.sum()
```

```python
joint_prob
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap='viridis')
ax.view_init(30, 45)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

```python
fig,ax=plt.subplots(1,1)
cp=ax.contourf(X,Y,Z)
fig.colorbar(cp)
ax.set_title('Contour Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')

```

```python
num_bins=25
num_samples=10000
```

```python
mean=[0.0,0.0]
cov =  [[1, 0.8], 
        [0.8, 1.0]] 
data= 100 + 10 * np.random.multivariate_normal(mean,cov,num_samples)
joint_prob,edges=np.histogramdd(data,num_bins)
edges=np.array(edges)[:,1:]
print(joint_prob)
```

```python
cut=14
X_Vals=np.array(joint_prob[cut])
X_Vals = X_Vals/X_Vals.sum()
Y_Vals=np.array([row[cut] for row in joint_prob])
Y_Vals = Y_Vals/Y_Vals.sum()
```

```python
X_Vals
```

```python
plt.bar(edges[0],X_Vals)
plt.xlabel('X Values')
plt.ylabel('Frequency')
plt.title('Distribution of X Values')
```

```python
plt.bar(edges[1],Y_Vals)
plt.xlabel('Y Values')
plt.ylabel('Frequency')
plt.title('Distribution of Y Values')
```

```python
X=edges[0]
Y=edges[1]
X,Y=np.meshgrid(X,Y)
Z= joint_prob/joint_prob.sum()
```

```python
fig=plt.figure(figsize=(9,6))
ax=plt.axes(projection='3d')
ax.view_init(30,300)
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')            
```

```python
fig,ax = plt.subplots(1,1)
cp = ax.contourf(X, Y, Z)
fig.colorbar(cp) # Add a colorbar to a plot
ax.set_title('Contour Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
```
