```python
class Vector:
    def __init__(self,x=0,y=0):
        self.x=x
        self.y=y
    def __str__(self):
        return "["+str(self.x)+","+str(self.y)+"]"
```

```python
a=Vector()
print(a)
```

```python
a=Vector(5,5)
```

```python
print(a)
```

```python
def add(self,b):
    c=Vector()
    c.x=self.x + b.x
    c.y=self.y + b.y
    return c
    
```

```python
Vector.add=add
```

```python
b=Vector(4,6)
d=a.add(b)
print(d)
```

```python
def sub(self,b):
    c=Vector()
    c.x=self.x - b.x
    c.y=self.y - b.y
    return c
Vector.sub=sub       
```

```python
c=a.sub(b)
print(c)
```

```python
def mul(self,b):
    c=Vector()
    c.x=self.x * b
    c.y=self.y * b
    return Vector(str(c.x),str(c.y))
Vector.mul=mul
```

```python
print(a.mul(5))
```

```python
def dot_product(self,b):
    c.x= self.x * b.y
    c.y= self.y * b.x
    return c.x + c.y
Vector.dot_product=dot_product
```

```python
d_p=a.dot_product(b)
print(d_p)
```

```python
class Matrix:
    def __init__(self,dim,fill): #Dimension will be Tuple and fill will the element from whose we want to create the list
        self.rows=dim[0]
        self.cols=dim[1]
        self.A=[                 #This is called as list comprehension.
                                 #Square brackets also the part of syntax.
            [fill] * self.cols   #fill * self.cols here results in [2.0 2.0 2.0] (Like the number of columns we indicate here)
            for i in range(self.rows)
        ] 
```

```python
def __str__(self):
    rows = len(self.A) # here len returns the number of elements it has as I have made a list of list earlier
    ret = ''
    for i in range(rows):
        cols = len(self.A[i]) #here it will move to 1st element(Considering its length there)
        for j in range(cols):
            ret += str(self.A[i][j]) + "\t"
        ret += "\n"
    return ret
Matrix.__str__=__str__
```

```python
class Matrix:
    def __init__(self,dim):
        self.rows=dim[0]
        self.cols=dim[1]
        self.vals={}    
    
```

```python
def set(self,i,j,val):
    if i<0 or i>self.rows:
         print("Rows index out of range")
    if j<0 or j>self.cols:
        print("Cols index out of range")
    self.vals[(i,j)]=val

def get(self,i,j):
    if i<0 or i>self.rows:
         print("Rows index out of range")
    if j<0 or j>self.cols:
        print("Cols index out of range")
    if (i,j) in self.vals:
        return self.vals[(i,j)]
    return 0.0
Matrix.set=set
Matrix.get=get
```

```python
import math
def basic_sigmoid(x): #This is writing sigmoid function which gives 0 at -infinity and 1 at +ve infinity.
    """
    Compute sigmoid of x. 

    Arguments:
    x -- A scalar

    Return:
    s -- sigmoid(x)
    """
    
    s = 1./(1. + math.e ** (-x))
    
    return s
```

```python
x=math.sqrt(25)
x
```

```python
basic_sigmoid(x)
```

```python
 z=[2,4,5,3] # It doesnot work because functions are defined to be work on scalar. Not lists
basic_sigmoid(z) 
```
