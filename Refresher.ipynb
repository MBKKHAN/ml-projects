```python
def good_enough(x,guess):
    return abs((guess*guess)-x)<0.00000000000001  #since squareroot => âˆšx=y2 and y2-x=0
```

```python
def improve_guess(x,guess):
    return (guess+x/guess)/2  # guess + x/guess gives mid value of guess and improved guess
    #like if its too big it will keep to small.
    #if its too small it will get it to big.
def sqrt(x,guess=1):
    if good_enough(x,guess):
        return guess
    else:
        return sqrt(x,improve_guess(x,guess))
```

# "OOPS Concept in Python"

```python
class Point:
    def __init__(self,x=0,y=0):
        self.x=x
        self.y=y
    def __str__(self):
        return "["+str(self.x)+","+str(self.y)+"]"
```

```python
p1=Point()
print(p1)
p2=(Point(4,5))
print(p2)
```

```python
class Shape:
    def __init__(self,points):
        self.points=points
    def __str__(self):
        ret=""
        for point in self.points:
            ret += str(point)+"->"
        return ret
```

```python
p1=Point(2,2)
p2=Point(4,2)
p3=Point(3,5)
p=[p1,p2,p3]
sh=Shape(p)
print(sh)
```

```python
def PrintPoints(self):
    for i in self.points:
        print(i)
Shape.PrintPoints=PrintPoints
```

```python
sh.PrintPoints()
```

```python
class Triangle(Shape):
    pass

t1=Triangle(p)
```

```python
def get_area(self):
    vertices=self.points
    n=len(vertices)
    a=0.0
    for i in range(n):
        j=(i+1)%n
        a += abs((vertices[i].x*vertices[j].y-vertices[j].x*vertices[i].y))/2
    return a
Triangle.get_area=get_area
```

```python
t1.get_area()
```

```python
class Rectangle:
    def __init__(self,length,width):
        self.length=length
        self.width=width
    def area(self):
        return self.length*self.width
    def perimeter(self):
        return 2*(self.length+self.width)
    def __str__(self):
        return "L: "+str(self.length)+", W: "+str(self.width)
    
```

```python
class Square(Rectangle):
    def __init__(self,length):
        super().__init__(length,length)
```

```python
rect=Rectangle(2,5)
print(rect.area())
```

```python
s=Square(5)
s.area()
s.perimeter()
```

```python
class Vector:
    def __init__(self,x=0.0,y=0.0):
        self.x=x
        self.y=y
    def __str__(self):
        return "[{},{}]".format(str(self.x),str(self.y))
```

```python
a=Vector(6.6,8)
print(a)
```

```python
def add(self,b):
    c=Vector()
    c.x=self.x + b.x
    c.y=self.y + b.y
    return c
```

```python
Vector.add=add
```

```python
b=Vector(5,6)
c=a.add(b)
```

```python
print(c)
```

```python
def sub(self,b):
    c=Vector()
    c.x=self.x - b.x
    c.y=self.y - b.y
    return c
```

```python
Vector.sub=sub
```

```python
c=a.sub(b)
```

```python
print(c)
```

```python
def mul(self,s):
    return Vector(s*self.x , s*self.y)
```

```python
Vector.mul=mul
```

```python
d=a.mul(5)
print(d)
```

```python
def sub(self,b):
    return self.add(b.mul(-1))
```

```python
v=Vector(5,6)
u=Vector(8,5)
```

```python
s=v.sub(u)
print(s)
```

```python
def cross_product(self,b):
    return Vector(self.x*b.y-b.x*self.y)
```

```python
Vector.cross_product=cross_product
```
